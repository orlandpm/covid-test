```python
def row_brightness(img):
    width, height = img.size
    out = []
    for i in range(height):
        row = [img.getpixel((j,i)) for j in range(width)]
        avg = sum(r+g+b for r,g,b in row)/len(row)
        out.append(avg)
    return out
        
    
def normalize(v):
    ma = max(v)
    mi = min(v)
    ys = [1-(y-mi)/(ma-mi) for y in v]
    l = len(v)
    xs = [i/l for i in range(l)]
    return xs,ys

def test_curve(img):
    return normalize(row_brightness(img))
```

```python
plt.plot(*test_curve(july9), label='July9')
plt.plot(*test_curve(july13), label='July13')
plt.plot(*test_curve(july14), label='July14')
ax = plt.subplot(1,1,1)
ax.legend()
```
